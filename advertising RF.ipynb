```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("D:/ML n DA\Machine learning project/Advertise/advertising.csv")
```

```python
X=df.iloc[:,:-1].values
y=df.iloc[:,3].values
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor2 = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor2.fit(X, y)


y_pred = regressor2.predict(X)
```

```python
print(y_pred.reshape(-1,1))
```

```python
regressor2.score(X,y)
```
